```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('E:\\archive\\games.csv')
df1
```

```python
df1.info()
```

```python
df1['Attendance'] = df1['Attendance'].str.replace(',', '', regex=False).replace('', '0').astype(int)
df1.info()
```

```python
missing_values = df1.isnull().sum()
missing_values
```

```python
missing_percentage = (df1.isnull().mean() * 100).round(2)
missing_percentage
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df1[['MaxTemp', 'MinTemp', 'Rainfall']])
plt.title('Distribution of Weather Variables')
plt.ylabel('Value')
plt.xlabel('Variable')
plt.show()
```

```python
df1.fillna({'MaxTemp': df1['MaxTemp'].median(),
            'MinTemp': df1['MinTemp'].median(),
            'Rainfall': 0}, inplace=True)
```

```python
import math

num_cols = df1.select_dtypes(include=['float64', 'int64' , 'int32']).columns
n_cols = 2
n_rows = math.ceil(len(num_cols) / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, n_rows * 4))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.histplot(df1[col], kde=True, bins=30, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, n_rows * 4))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.boxplot(y=df1[col], ax=axes[i])
    axes[i].set_title(f'Boxplot of {col}')
    axes[i].set_ylabel(col)

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
for col in df1.select_dtypes(include=['float64', 'int64' , 'int32']).columns:
    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df1[col] = df1[col].clip(lower, upper)
```

```python
import math

num_cols = df1.select_dtypes(include=['float64', 'int64']).columns
n_cols = 2
n_rows = math.ceil(len(num_cols) / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, n_rows * 4))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.boxplot(y=df1[col], ax=axes[i])
    axes[i].set_title(f'Boxplot of {col} (Outliers Capped)')
    axes[i].set_ylabel(col)

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 10))
corr = df1[num_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
